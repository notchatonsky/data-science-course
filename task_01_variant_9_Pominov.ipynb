```python
# =========================================
# ШИФР ЦЕЗАРЯ СО СДВИГОМ НА 3
# =========================================

# 1. Запрашиваем у пользователя сообщение для шифрования
original_message = input("Введите сообщение для шифрования: ")

# 2. Определяем тип данных, с которым будем работать
import string

# Полный набор символов, которые мы будем сдвигать.
rus_lower = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
rus_upper = 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
eng_lower = string.ascii_lowercase  # 'abcdefghijklmnopqrstuvwxyz'
eng_upper = string.ascii_uppercase  # 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# 3. Задаем сдвиг
shift = 3

# Функция для шифрования/дешифрования текста
def caesar_cipher(text, shift_amount):
    """
    Шифрует или дешифрует текст, сдвигая буквы и цифры.
    Регистр букв сохраняется.
    """
    result = []
    
    for char in text:
        # Шифрование для русских букв (нижний регистр)
        if char in rus_lower:
            index = rus_lower.index(char)
            new_index = (index + shift_amount) % len(rus_lower)
            result.append(rus_lower[new_index])
        # Шифрование для русских букв (верхний регистр)
        elif char in rus_upper:
            index = rus_upper.index(char)
            new_index = (index + shift_amount) % len(rus_upper)
            result.append(rus_upper[new_index])
        # Шифрование для английских букв (нижний регистр)
        elif char in eng_lower:
            index = eng_lower.index(char)
            new_index = (index + shift_amount) % len(eng_lower)
            result.append(eng_lower[new_index])
        # Шифрование для английских букв (верхний регистр)
        elif char in eng_upper:
            index = eng_upper.index(char)
            new_index = (index + shift_amount) % len(eng_upper)
            result.append(eng_upper[new_index])
        # Если символ не буква, оставляем без изменений
        else:
            result.append(char)
            
    return ''.join(result)

# 4. Выполняем шифрование и дешифрование
encrypted_message = caesar_cipher(original_message, shift)
decrypted_message = caesar_cipher(encrypted_message, -shift)

# 5. Выводим результаты
print("\n" + "=" * 50)
print("РЕЗУЛЬТАТЫ ШИФРОВАНИЯ")
print("=" * 50)
print(f"Исходное сообщение:     {original_message}")
print(f"Зашифрованное сообщение: {encrypted_message}")
print(f"Расшифрованное сообщение: {decrypted_message}")
print("=" * 50)

# Проверка, что расшифровка прошла успешно
if original_message == decrypted_message:
    print("Проверка пройдена: расшифрованное сообщение совпадает с исходным.")
else:
    print("Ошибка: расшифрованное сообщение не совпадает с исходным.")
print("=" * 50)
```
